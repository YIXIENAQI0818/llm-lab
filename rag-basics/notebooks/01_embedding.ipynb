# 01 Embedding 基础

把文本变成向量，语义相近的文本向量也相近。

```python
import sys
sys.path.insert(0, '..')

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

from src.embed import embed
```

## 文本 → 向量

```python
sentences = [
    "今天天气真好，阳光明媚。",
    "今天是个大晴天，非常适合出门。",
    "我喜欢吃披萨和意大利面。",
    "意大利菜是我最爱，尤其是披萨。",
    "深度学习是人工智能的重要分支。",
]

vectors = embed(sentences)

print(f"句子数: {vectors.shape[0]}, 向量维度: {vectors.shape[1]}")
print(f"每个值都是 float, 范围约在 [{vectors.min():.3f}, {vectors.max():.3f}]")
```

## 语义相似度

```python
sim = cosine_similarity(vectors)

for i, s in enumerate(sentences):
    for j in range(i+1, len(sentences)):
        print(f"[{sim[i][j]:.3f}] {s}  ←→  {sentences[j]}")
    print()
```
