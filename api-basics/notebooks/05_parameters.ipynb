# 05 参数调优

temperature、top_p、max_tokens 如何影响模型输出。

```python
import sys
sys.path.insert(0, '..')

from src.client import get_client

client = get_client()
```

## Temperature — 创造力高低

```python
prompt = "用一个比喻描述编程这件事。"

for temp in [0, 1, 1.5]:
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[{"role": "user", "content": prompt}],
        temperature=temp,
    )
    print(f"temperature={temp}")
    print(response.choices[0].message.content)
    print("---")
```

## top_p — 候选词范围

```python
prompt = "随机想一句送给朋友的祝福语。"

for p in [0.1, 0.5, 0.95]:
    print(f"=== top_p={p} ===")
    for i in range(3):
        response = client.chat.completions.create(
            model="deepseek-chat",
            messages=[{"role": "user", "content": prompt}],
            top_p=p,
        )
        print(f"  第{i+1}次: {response.choices[0].message.content}")
    print()
```

## max_tokens — 回复长度上限

```python
prompt = "请详细介绍深度学习的发展历史。"

for limit in [30, 100, 200]:
    response = client.chat.completions.create(
        model="deepseek-chat",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=limit,
    )
    print(f"max_tokens={limit}")
    print(response.choices[0].message.content)
    print("---")
```
