## 分组 1：相关与无关混合

```python
reranker = Reranker()

candidates = [
    {"score": 0.90, "text": "AHP法是经典的多属性决策方法，通过两两比较确定各准则权重，能系统化整理专家经验", "meta": {"name": "chunk0"}},
    {"score": 0.85, "text": "TOPSIS法通过计算各方案与理想解的距离来进行排序", "meta": {"name": "chunk1"}},
    {"score": 0.80, "text": "目标威胁评估是现代战争中的重要环节，需要综合考虑距离、速度、机动能力等因素", "meta": {"name": "chunk2"}},
    {"score": 0.75, "text": "Transformer架构由Google在2017年提出，基于Self-Attention机制", "meta": {"name": "chunk3"}},
    {"score": 0.70, "text": "北京今天天气晴朗，气温25度", "meta": {"name": "chunk4"}},
]

query = "如何用AHP法构建目标威胁评估指标体系"

print(f"查询: {query}\n")
print("=== 粗排结果 (RRF) ===")
for i, c in enumerate(candidates):
    print(f"{i+1}. [{c['meta']['name']}] rrf={c['score']:.3f}")

reranked = reranker.rerank(query, candidates, top_k=5)

print("\n=== Cross-Encoder 精排 ===")
for i, c in enumerate(reranked):
    print(f"{i+1}. [{c['meta']['name']}] ce={c['score']:.4f} {c['text'][:50]}")
```
